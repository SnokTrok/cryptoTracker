# RealTime crypto currency value from Binance using Websocket.

- Using websockt api for binance to pull the real time data for supplied crypto's
- This would be a perpetual process launched on the cloud somewhere ideally

```python
import websocket
import pandas  as pd
import json
```

```python
# currency websocket endpoints to subscribe to 
subscribed = ['WBTCUSDT', 'WBTCETH']

# kline_1m -> minuitely updated stream of data 
subscribed = "/".join([coin.lower() + '@kline_1m' for coin in subscribed])
```

We can use these kline websocket enpoints to gatehr data on the exchange prices BETWEEN different cryptoCurrencys, 

THIS IS NOT NECESSARILY 1:1 with USD , the USDT coin is very close representation to this but it is not guaranteed

```python

socket = "wss://stream.binance.com:9443/stream?streams=" + subscribed
    
```

```python
def on_binance_kline_message(ws,message)-> pd.DataFrame:
    print(f"Recieved binance websocket stream : ")
    data = json.loads(message)['data']
    
    # check kline closed
    if data['k']['x'] == False:
        print("Candle not closed , skipping...")
        return None
    
    df = pd.DataFrame.from_records([data['k']])
    df=df.rename({
        't' : 'date_open',
        'T' : 'date_close',
        "s" : 'symbol',
        'i' : 'interval',
        'o' : 'price_open',
        'c' : 'price_close',
        'h' : 'price_high',
        'l' : 'price_low',
        'n' : 'num_trades'
    },axis=1)
    df=df.astype({
        "symbol" : 'string',
        'interval' : 'string',
        'price_open' : 'Float64',
        'price_close' :'Float64',
        'price_high' : 'Float64',
        'price_low' : 'Float64',
        'num_trades' : 'Int64'
    })

    parse_epoch  = ['date_open' , 'date_close']

    for col in parse_epoch:
        df[col] = df[col].apply(lambda x : pd.to_datetime(x).replace(tzinfo=None))

    
```

```python
ws = websocket.WebSocketApp(socket,on_message=on_binance_kline_message) 
websocket.enableTrace(True)
ws.run_forever()
```
