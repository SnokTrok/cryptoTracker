```python
# @api_request
def __get_latest_block_number() -> str:
    """
        Returns HexId of latest block number , can convert into int using int(i , 16)
    """
    print(f"Getting latest block number as  hex str")
    payload = {
        "jsonrpc": "2.0",
        "id": 1,
        "method": "eth_blockNumber",  # Replace with the method you want to call
        "params": []  # Replace with the appropriate parameters
    }
    resp = requests.post(endpoint,headers={'Content-Type' : 'application/json'},data=json.dumps(payload))
    assert resp.status_code == 200
    return resp.json()['result']

resp = __get_latest_block_number()
resp
```

```python
def get_block(blockId : str , full_transactions : bool = True) -> dict:
    """
        Returns block information including trsaction history optionally full or hashed
    """
    payload = {
        "jsonrpc": "2.0",
        "id": 1,
        "method": "eth_getBlockByNumber",  # Replace with the method you want to call
        "params": [
            blockId,full_transactions
        ] 
    }
    resp = requests.post(endpoint,headers={'Content-Type' : 'application/json'},data=json.dumps(payload))
    assert resp.status_code == 200
    return resp.json()
```

```python
import pandas as pd
def convert_eth_history_to_df(data : dict)-> pd.DataFrame:
    date_cols = ['date_open','date_close']
    dtype = {
        'price_open' : 'Float64',
        'price_close' : 'Float64',
        'price_high' : 'Float64',
        'price_low' : 'Float64',
        'volume' : 'Float64',
        'num_trades' : 'Int64',
       # 'chain' : 'string',
        #'token_name' : 'string'
    }
    df = pd.DataFrame.from_records(data).drop(['timestamp_open','timestamp_close'],axis=1)
    df = df.astype(dtype)

    for col in date_cols:
        df[col] = df[col].apply(lambda x : pd.to_datetime(x).replace(tzinfo=None))


    return df
```
